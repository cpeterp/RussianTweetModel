```python
import pandas as pd
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split as tts
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB         # Naive Bayes
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from textblob import TextBlob, Word
from nltk.stem.snowball import SnowballStemmer

%matplotlib inline
```

```python
tweet_data = pd.read_csv("./datasets/modified_data/combined_tweets.csv", sep=";", 
                         dtype={'tweet_id':str, 'author_id':str, 'publish_date':str, 
                                'content':str, 'link_url':str, 'account_category':str, 
                                'author':str, 'account_type':str})
```

```python
tweet_sample = tweet_data.sample(n=100000, random_state=42)
```

```python
sns.heatmap(tweet_sample.isnull(), yticklabels = False, cbar = False, cmap = "viridis")
```

```python
tweet_sample = tweet_sample[(tweet_sample['account_category'] == "Politician") | (tweet_sample['account_category'] == "Troll")]
```

```python
plt.subplots(figsize=(10,6))
sns.countplot(x=tweet_sample['account_category'])
```

```python
tweet_sample.loc[(tweet_sample['account_type'] == "Democrat"), 'account_type'] = "Left"
```

```python
tweet_sample.loc[(tweet_sample['account_type'] == "Republican"), 'account_type'] = "Right"
```

```python
plt.subplots(figsize=(10,6))
sns.countplot(x=tweet_sample['account_type'])
```

```python
plt.subplots(figsize=(10,6))
sns.countplot(x=tweet_sample['account_category'], hue=tweet_sample['account_type'])
```

```python
tweet_sample = pd.get_dummies(tweet_sample, columns=['account_type', 'account_category'], drop_first=True)
```

```python
tweet_sample.columns
```

```python
# Define a function that accepts text and returns the polarity.
def detect_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
tweet_sample['sentiment'] = tweet_sample.content.apply(detect_sentiment)
```

```python
tweet_sample.boxplot(column='sentiment', by='account_category_Troll')
```

```python
tweet_sample.boxplot(column='sentiment', by='account_type_Right')
```

```python
tweet_sample['length'] = tweet_sample['content'].apply(lambda x: len(x))
```

```python
tweet_sample.boxplot(column='length', by='account_category_Troll')
```

```python
tweet_sample.boxplot(column='length', by='account_type_Right')
```
