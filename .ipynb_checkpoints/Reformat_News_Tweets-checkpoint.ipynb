```python
import pandas as pd
import json

data_columns=['tweet_id','author_id','publish_date','content','link_url','account_category','author','account_type']
news_handles = pd.read_csv("datasets_large/news-tweets/new_outlets-accounts.csv")
print(news_handles)
```

```python
# Data is in JSON format, need to read in data and only save ones in the selected accounts given by new_handles
i = 0
tweet_data = pd.DataFrame(columns=data_columns)
with open("datasets_large/news-tweets/news_tweets.txt", encoding="UTF-8") as file: # Use file to refer to the file object
    for line in file:
        i += 1
        if i%10000==0:
            print("Lines read: {}".format(i))
            print()
        line = json.loads(line)
        author = line['user']['screen_name']
        if news_handles['handle'].isin([author]).any():
            i += 1
            line_dict = {
                'tweet_id' : line['id'],
                'author_id' : line['user']['id_str'],
                'publish_date' : line['created_at'],
                'content' : line['full_text'],
                'account_category' : "US_News",
                'author' : author,
                'account_type' : news_handles.loc[news_handles['handle'] == author, 'Lean'].item()   
            }
            tweet_data = tweet_data.append(line_dict, ignore_index=True)
            
```

```python
tweet_data.to_csv("datasets_large/news-tweets/news_tweets.csv", index=False, sep=";")
```
